# Summary

- This project aims to find disease outbreaks from synthetic news headlines. These headlines contain city and/or country mentions along with a particular disease such as "Zika hits Miami Beach" and "Bejing is expecting an outbreak of Hepatitis B" and so on. Given these headlines, the goal is to map out disease outbreaks to tightly connected geographic clusters.

- **Methodology**:
    - To extract locations I used regular expressions and the `geonamescache` library.
    - To map location mentions to their corresponding longitude and latitudes, I used `geonamescashe` again. Cities the same name were tie broken based on population.
    - To cluster the locations, I used KMeans and DBSCAN algorithms using euclidean as well as great circle distance. The parameters were chosen by trial and error. I found DBSCAN to be produce the most useful in this case.
    - To find disease outbreaks, I found the centers of each cluster and dug around the headlines in that cluster to find repeated diseases.
- **Output**:
    - I find that _Zika_ virus to be the most widespread outbreak spanning the south east part of the US, Central America, Easter South America, and south Asia. In addition, the outbreak of _Mad Cow Disease_ seems to concentrate in Europe.
